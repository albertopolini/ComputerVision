# Imports

```python
import numpy as np

# Torch imports
import torch
import torchvision


# NN Layers
from torch.nn import Conv2d, MaxPool2d, AdaptiveAvgPool2d

# Activation Functions
import torch.functional as FF

# Dataset import 
from torchvision import datasets

# Image Transformations
import torchvision.transforms as transforms

# Data Visualization
import matplotlib.pyplot as plt
```

# Loading the Dataset

```python
# Loading the dataset in the data folder

transformations = transforms.ToTensor()

cifar_trainset = datasets.CIFAR10(root='../Data/CIFAR10', train=True, download=True, transform=transformations)
cifar_testset = datasets.CIFAR10(root='../Data/CIFAR10', train=False, download=True, transform=transformations)
```

## Defining the Dataloader

```python
BATCH_SIZE = 32
```

```python
trainloader = torch.utils.data.DataLoader(cifar_trainset, batch_size=BATCH_SIZE,
                                          shuffle=True, num_workers=2)

testloader = torch.utils.data.DataLoader(cifar_testset, batch_size=BATCH_SIZE,
                                         shuffle=False, num_workers=2)

classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
def imshow(img):
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()


# get some random training images
dataiter = iter(trainloader)
images, labels = next(dataiter)

# show images
imshow(torchvision.utils.make_grid(images))
# print labels
print(' '.join(f'{classes[labels[j]]:5s}' for j in range(BATCH_SIZE)))
```
